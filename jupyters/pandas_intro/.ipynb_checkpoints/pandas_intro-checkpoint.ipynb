```python
import pandas
df7 = pandas.read_csv("supermarkets.csv")
df7
```

```python
df7["Address"] = df7["Address"] + ", " + df7["City"] + ", " + df7["State"] + ", " + df7["Country"]
df7
```

```python
# .apply is a pandas method
df7["Coordinates"] = df7["Address"].apply(nom.geocode)
df7
```

```python
# Use a lambda function
df7["Latitude"] = df7["Coordinates"].apply(lambda x: x.latitude if x != None else None)
df7["Longitude"] = df7["Coordinates"].apply(lambda x: x.longitude if x != None else None)
df7
```
